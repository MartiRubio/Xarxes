```python
#data
one_meter = [-52, -50, -51, -54, -54, 
             -56, -54, -54, -54, -52,
             -55, -53, -54, -52, -51, 
             -54, -58, -61, -56, -55]

two_meters = [-56, -59, -57, -57, -61, 
              -61, -55, -57, -59, -59,
              -60, -62, -62, -56]

three_meters = [-61, -70, -65, -74, -71,
                -64, -61, -70, -63, -74,
                -61, -67, -67, -64, -62,
                -65, -70, -60, -64, -65]

four_meters = [-71, -63, -72, -73, -71, 
               -78, -68, -72, -68, -69, 
               -64, -68]
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
plt.hist(one_meter)
plt.show()
plt.hist(two_meters)
plt.show()
plt.hist(three_meters)
plt.show()
plt.hist(four_meters)
plt.show()
```

```python
one = pd.DataFrame(data=one_meter, columns=['dB'])
two = pd.DataFrame(data=two_meters, columns=['dB'])
three = pd.DataFrame(data=three_meters, columns=['dB'])
four = pd.DataFrame(data=four_meters, columns=['dB'])
one['distance'] = 1
two['distance'] = 2
three['distance'] = 3
four['distance'] = 4
total = pd.concat([one, two, three, four])
total.index=[i for i in range(1, len(total) + 1)]
total
```

```python
plt.hist(total[total['distance'] == 1], bins = 2, label='x', alpha=0.4)
plt.hist(total[total['distance'] == 2], bins = 2,label='y', alpha=0.4)
plt.hist(total[total['distance'] == 3], bins = 2,label='z', alpha=0.4)
plt.show()
```
