# Entrega Codificació

```python
import random
import time
import matplotlib.pyplot as plt
import numpy
```

```python
bits = [random.randint(0,1) for i in range(10)]
print("Cadena original: ", bits)
fig = plt.figure()
ax = fig.gca()
ax.set_xticks(numpy.arange(0, 10, 1))
plt.step(range(10), bits)
plt.grid(axis='x')
plt.ylim(-1.5,1.5)
plt.show()
```

## Non return to zero

```python
def nonreturn_zero(bits=[0 for i in range(10)]):
    print("Codificació Non-Return-Zero")
    time.sleep(1)
    print("Senyal original: ")
    fig = plt.figure()
    ax = fig.gca()
    ax.set_xticks(numpy.arange(0, 10, 1))
    plt.step(range(10), bits)
    plt.grid(axis='x')
    plt.ylim(-.5,1.5)
    plt.show()
    time.sleep(1)
    cod = []
    for i in bits:
        if i:
            cod.append(1)
        else:
            cod.append(0)
    print("Senyal codificada: ")
    fig = plt.figure()
    ax = fig.gca()
    ax.set_xticks(numpy.arange(0, 10, 1))
    plt.step(range(10), cod)
    plt.grid(axis='x')
    plt.ylim(-.5,1.5)
    plt.show()
    time.sleep(1)
    decod = []
    for i in cod:
        if i:
            decod.append(1)
        else:
            decod.append(0)
    print("Senyal decodificada")
    fig = plt.figure()
    ax = fig.gca()
    ax.set_xticks(numpy.arange(0, 10, 1))
    plt.step(range(10), decod)
    plt.grid(axis='x')
    plt.ylim(-.5,1.5)
    plt.show()
    print("\nDIFF: ", [bits[i] - decod[i] for i in range(10)], "\n")
nonreturn_zero(bits)
```

## Non return to zero level

```python
def nonreturn_zero_level(bits=[0 for i in range(10)]):
    print("Codificació Non-Return to Zero Level")
    time.sleep(1)
    print("Senyal original: ")
    fig = plt.figure()
    ax = fig.gca()
    ax.set_xticks(numpy.arange(0, 10, 1))
    plt.step(range(10), bits)
    plt.ylim(-.5,1.5)
    plt.grid(axis='x')
    plt.show()
    time.sleep(1)
    cod = []
    for i in bits:
        if not i:
            cod.append(1)
        else:
            cod.append(0)
    print("Senyal codificada: ")
    fig = plt.figure()
    ax = fig.gca()
    ax.set_xticks(numpy.arange(0, 10, 1))
    plt.step(range(10), cod)
    plt.grid(axis='x')
    plt.ylim(-.5,1.5)
    plt.show()
    time.sleep(1)
    decod = []
    for i in cod:
        if not i:
            decod.append(1)
        else:
            decod.append(0)
    print("Senyal decodificada")
    fig = plt.figure()
    ax = fig.gca()
    ax.set_xticks(numpy.arange(0, 10, 1))
    plt.step(range(10), decod)
    plt.grid(axis='x')
    plt.ylim(-.5,1.5)
    plt.show()
    print("\nDIFF: ", [bits[i] - decod[i] for i in range(10)], "\n")
    
nonreturn_zero_level(bits)
```

## Non return to zero inverted

```python
def nonreturn_zero_inverted(bits=[0 for i in range(10)]):
    print("Codificació Non-Return Zero Inverted")
    time.sleep(1)
    print("Senyal original: ")
    fig = plt.figure()
    ax = fig.gca()
    ax.set_xticks(numpy.arange(0, 10, 1))
    plt.step(range(10), bits)
    plt.ylim(-.5,1.5)
    plt.grid(axis='x')
    plt.show()
    time.sleep(1)
    cod = []
    for i in range(10):
        if i == 0:
            if bits[i]:
                cod.append(1)
            else:
                cod.append(0)
        else:
            if bits[i]:
                cod.append(not cod[i-1])
            else:
                cod.append(cod[i-1])
    print("Senyal codificada: ")
    fig = plt.figure()
    ax = fig.gca()
    ax.set_xticks(numpy.arange(0, 10, 1))
    plt.step(range(10), cod)
    plt.ylim(-.5,1.5)
    plt.grid(axis='x')
    plt.show()
    time.sleep(1)
    decod = []
    for i in range(10):
        if i == 0:
            if cod[i]:
                decod.append(1)
            else:
                decod.append(0)
        else:
            if cod[i] == cod[i-1]:
                decod.append(0)
            else:
                decod.append(1)
    print("Senyal decodificada")
    fig = plt.figure()
    ax = fig.gca()
    ax.set_xticks(numpy.arange(0, 10, 1))
    plt.step(range(10), decod)
    plt.ylim(-.5,1.5)
    plt.grid(axis='x')
    plt.show()
    print("\nDIFF: ", [bits[i] - decod[i] for i in range(10)], "\n")
    
nonreturn_zero_inverted(bits)
```

## Bipolar AMI

```python
def bipolar_ami(bits=[0 for i in range(10)]):
    print("Codificació Bipolar AMI")
    time.sleep(1)
    print("Senyal original: ")
    fig = plt.figure()
    ax = fig.gca()
    ax.set_xticks(numpy.arange(0, 10, 1))
    plt.step(range(10), bits)
    plt.ylim(-.5,1.5)
    plt.grid(axis='x')
    plt.show()
    time.sleep(1)
    cod = []
    last_high = False
    for i in bits:
        if i:
            if last_high:
                cod.append(-1)
                last_high = not last_high
            else:
                cod.append(1)
                last_high = not last_high
        else:
            cod.append(0)
    print("Senyal codificada: ")
    fig = plt.figure()
    ax = fig.gca()
    ax.set_xticks(numpy.arange(0, 10, 1))
    plt.step(range(10), cod)
    plt.ylim(-1.5,1.5)
    plt.grid(axis='x')
    plt.show()
    time.sleep(1)
    decod = []
    for i in cod:
        if i:
            decod.append(1)
        else:
            decod.append(0)
    print("Senyal decodificada")
    fig = plt.figure()
    ax = fig.gca()
    ax.set_xticks(numpy.arange(0, 10, 1))
    plt.step(range(10), decod)
    plt.ylim(-1.5,1.5)
    plt.grid(axis='x')
    plt.show()
    print("\nDIFF: ", [bits[i] - decod[i] for i in range(10)], "\n")
    
bipolar_ami(bits)
```

## Pseudoternary

```python
def pseudoternary(bits=[0 for i in range(10)]):
    print("Codificació Pseudoternary")
    time.sleep(1)
    print("Senyal original: ")
    fig = plt.figure()
    ax = fig.gca()
    ax.set_xticks(numpy.arange(0, 10, 1))
    plt.step(range(10), bits)
    plt.ylim(-.5,1.5)
    plt.grid(axis='x')
    plt.show()
    time.sleep(1)
    cod = []
    last_high = False
    for i in bits:
        if not i:
            if last_high:
                cod.append(-1)
                last_high = not last_high
            else:
                cod.append(1)
                last_high = not last_high
        else:
            cod.append(0)
    print("Senyal codificada: ")
    fig = plt.figure()
    ax = fig.gca()
    ax.set_xticks(numpy.arange(0, 10, 1))
    plt.step(range(10), cod)
    plt.ylim(-1.5,1.5)
    plt.grid(axis='x')
    plt.show()
    time.sleep(1)
    decod = []
    for i in cod:
        if i:
            decod.append(0)
        else:
            decod.append(1)
    print("Senyal decodificada")
    fig = plt.figure()
    ax = fig.gca()
    ax.set_xticks(numpy.arange(0, 10, 1))
    plt.step(range(10), decod)
    plt.ylim(-.5,1.5)
    plt.grid(axis='x')
    plt.show()
    print("\nDIFF: ", [bits[i] - decod[i] for i in range(10)], "\n")
    
pseudoternary(bits)
```

## Manchester

```python
def manchester(bits=[0 for i in range(10)]): 
    print("Codificació Manchester")
    time.sleep(1)
    print("Senyal original: ")
    fig = plt.figure()
    ax = fig.gca()
    ax.set_xticks(numpy.arange(0, 10, 1))
    plt.step(range(10), bits)
    plt.ylim(-.5,1.5)
    plt.grid(axis='x')
    plt.show()
    time.sleep(1)
    cod = []
    for i in bits:
        if i:
            cod.append(0)
            cod.append(1)
        else:
            cod.append(1)
            cod.append(0)
    print("Senyal codificada: ")
    fig = plt.figure()
    ax = fig.gca()
    plt.step(range(20), cod)
    ax.set_xticks(numpy.arange(0, 20, 2))
    plt.grid(axis='x')
    plt.ylim(-.5,1.5)
    plt.show()
    time.sleep(1)
    decod = []
    for i in range(10):
        if cod[2*i] == 0:
            decod.append(1)
        else:
            decod.append(0)
    print("Senyal decodificada")
    fig = plt.figure()
    ax = fig.gca()
    plt.step(range(10), decod)
    ax.set_xticks(numpy.arange(0, 10, 1))
    plt.grid(axis='x')
    plt.ylim(-0.5,1.5)
    plt.show()
    print("\nDIFF: ", [bits[i] - decod[i] for i in range(10)], "\n")
    
manchester(bits)
```

## Differential Manchester

```python
def diff_manchester(bits=[0 for i in range(10)]):
    print("Codificació Differential Manchester")
    time.sleep(1)
    print("Senyal original: ")
    fig = plt.figure()
    ax = fig.gca()
    ax.set_xticks(numpy.arange(0, 10, 1))
    plt.step(range(10), bits)
    plt.ylim(-.5,1.5)
    plt.grid(axis='x')
    plt.show()
    time.sleep(1)
    cod = []
    start = True
    for i in bits:
        if start:
            if i:
                start = False
                cod.append(0)
                cod.append(1)
            else:
                cod.append(1)
                cod.append(0)
        else:
            if i:
                cod.append(not cod[len(cod) - 1])
                cod.append(not cod[len(cod) - 1])
            else:
                cod.append(cod[len(cod) - 1])
                cod.append(not cod[len(cod) - 1])
    print("Senyal codificada: ")
    fig = plt.figure()
    ax = fig.gca()
    ax.set_xticks(numpy.arange(0, 20, 2))
    plt.step(range(20), cod)
    plt.ylim(-.5,1.5)
    plt.grid(axis='x')
    plt.show()
    time.sleep(1)
    decod = []
    for i in cod:
        if i == 'transition':
            decod.append(1)
        else:
            decod.append(0)
    print("Senyal decodificada")
    fig = plt.figure()
    ax = fig.gca()
    ax.set_xticks(numpy.arange(0, 10, 1))
    plt.step(range(10), decod)
    plt.ylim(-.5,1.5)
    plt.grid(axis='x')
    plt.show()
    print("\nDIFF: ", [bits[i] - decod[i] for i in range(10)], "\n")
    
diff_manchester(bits)
```
